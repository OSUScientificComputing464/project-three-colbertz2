I'm having a hard time understanding what the information in `data` represents, so let's visualize some of the `images` data and see what that looks like.

```python
plt.imshow(data.images[0])
```

Okay, this seems to show that the `images` array contains 1797 images at 8x8 pixels each. Let's see some more.

```python
fig, ax = plt.subplots(4, 10, figsize=(15, 5))
fig.tight_layout()

for i, axi in enumerate(ax.ravel()):
    axi.imshow(data.images[i])
```

Okay, so the images are not necessarily in order from 0 to 9, but the digits seem to only include 0 to 9. The 8x8 images seem to indicate that the data is a flattened version of each image.

```python
print data.images[0].ravel()
print data.data[0]
```

So, the data array is definitely a flattened version of each image. The target array must be the classification of each image.

```python
fig, ax = plt.subplots(4, 10, figsize=(15, 5))
fig.tight_layout(pad=2)

for i, axi in enumerate(ax.ravel()):
    axi.imshow(data.images[i])
    axi.set_title(data.target[i])
```
